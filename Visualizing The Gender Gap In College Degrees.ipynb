# Visualizing The Gender Gap In College Degrees

# Introduction

This is a guided project provided by DataQuest.io. Its aim is to practice the basic data visualization technics in Python and introduce an improvement to the plot aesthetics.

This project is based on data from NCES 2013 Digest of Education Statistics, that was cleaned and prepared by Randy Olson on his private webpage (http://www.randalolson.com/2014/06/14/percentage-of-bachelors-degrees-conferred-to-women-by-major-1970-2012/).
The given data set contains the annual percentages of bachelor's degrees granted to women from 1970 to 2012. The database is divided into 17 categories of degrees. 

The first code below represents the result of my work throughout the "Storytelling Through Data Visualization" course. It is a visual analysis of the gender gap at the university. Data represent the percentage of granted bachelor's degrees to women in the USA between years 1970-2012 in STEM majors (what stands for academic disciplines related to Science, Technology, Engineering, and Mathematics).

This project was aimed to help people understand the gender gap in different fields across time. Following the principle set forth by Edward Tufte, a pioneer of the field of data visualization I wanted to create graphs without excess elements, known also as chartjunk. In other words, the goal is to maximize the data-ink ratio, which is the fractional amount of the plotting area dedicated to displaying the data. "Non-data ink includes any elements in the chart that don't directly display data points. This includes tick markers, tick labels, and legends. Data ink includes any elements that display and depend on the data points underlying the chart." Nevertheless, it is important to be mindful of the trade-off between non-data in and data-ink, as working on tweaking the appearance of plots to tell a story, because plots that are created could end up telling the wrong story.

I have set colors for the line plots based on the Tableau Color Blind 10 palette, to make sure that it will be readible to everyone. (" Approximately 8% of men and 0.5% of women of Northern European descent suffer from red-green color blindness.")

The charts here are ordered by decreasing the ending gender gap. I've populated the list stem_cats with the six STEM degree categories, ordering them by decreasing the ending gender gap. 

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

At first glance, we see that the biggest gender difference appears in the Engineering and Computer Science, where the percentage of granted degrees to women stays in minority during the whole period of study. It is maintained below 25% in the Engineering category and 40% in Computer Science. The smallest gender gap is visible in the Math and Statistics major, nevertheless, man preserves the dominant number in all years 1970-2012. " This chart can be improved and this project will explore techniques to do so. The previously mentioned Edward Tufte teaches that a good chart promotes comparison over simple descriptions. This implies using a consistent style for the components that aren't directly conveying the data points. This refers to the non-data ink used in the chart. Thanks to the maintenance of consistency in non-data ins across different plots, differences between those components are easier to be spotted by the viewer. Our visual processing systems work then better and are prone to faster identifying of the differences and bringing them on the first plan of our thought process.

# Visual representation of all degrees

 In this section, I will combine all the degrees that are included in the data set and I will generate line charts to compare across all degree categories. As I have mentioned in the project Introduction, there are 17 categories of degrees in our database. I have divided them into three genres: STEM category (6 majors), liberal arts(5 majors) and others (6 majors). They are saved as lists, where each of them is ordered descending by the percentage of degrees awarded to women.

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(16, 20))

## Generate first column of line charts. STEM degrees.##
for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if cat_index == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif cat_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')
        
## Generate second column of line charts. Liberal arts degrees.##
for sp in range(1,16,3):
    cat_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if cat_index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
  
 ## Generate third column of line charts. Other degrees.
for sp in range(2,20,3):
    cat_index = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if cat_index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif cat_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        
plt.show()
```

Having that many plots in one diagram create a sense of chaos while non-data components seem to clutter the view. The viewer might get overwhelmed trying to analyze the charts. The first issue that caught my eye is the fact that the titles of the singular plots are overlapping with the x-axis of the plots situated just above them. I cannot remove the title of the charts as the carry important information for identifying majors. However, I may remove the x-axis labels for the charts that have another chart located underneath them.

```python
fig = plt.figure(figsize=(16, 16))

## Generate first column of line charts. STEM degrees. ##

for sp in range(0,18,3):
    cat_index = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[cat_index]], c=cb_dark_blue, label="Women", linewidth=3)
    ax.plot(women_degrees["Year"], 100-women_degrees[stem_cats[cat_index]], c=cb_orange, label="Men", linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if cat_index == 0:
        ax.text(2005, 85, "Men")
        ax.text(2002, 10, "Women")
    elif cat_index == 5:
        ax.text(2005, 87, "Men")
        ax.text(2001, 7, "Women")
        ax.tick_params(labelbottom=True)
        
## Generate second column of line charts. Liberal arts degrees. ##
    
for sp in range(1,16,3):
    cat_index = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[cat_index]], c=cb_dark_blue, label="Women", linewidth=3)
    ax.plot(women_degrees["Year"], 100-women_degrees[lib_arts_cats[cat_index]], c=cb_orange, label="Men", linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if cat_index == 0:
        ax.text(2005, 78, "Men")
        ax.text(2002, 18, "Women")
    elif cat_index == 4:
        ax.tick_params(labelbottom=True)
  
 ## Generate third column of line charts. Other degrees. ##

for sp in range(2,20,3):
    cat_index = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees["Year"], women_degrees[other_cats[cat_index]], c=cb_dark_blue, label="Women", linewidth=3)
    ax.plot(women_degrees["Year"], 100-women_degrees[other_cats[cat_index]], c=cb_orange, label="Men", linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    
    if cat_index == 0:
        ax.text(2005, 90, "Men")
        ax.text(2002, 5, "Women")
    elif cat_index == 5:
        ax.text(2005, 62, "Men")
        ax.text(2001, 30, "Women")
        ax.tick_params(labelbottom=True)
        
plt.show()
```

The figure above looks much neater. I have kept the x-axis labels of the bottommost plots for the reference but removed them from all the plots above those. It is true that such a change will influence the ability of the viewer to determine the exact year of the given data point. However, the aim of this graph is to enable the viewer to get a big picture of the gender imbalance across different majors and quickly identify the trends and changes across the years.

Another step would be to declutter the y-axis labels. Since they are consistent across all the plots I can simplify them. At this point, I am not interested in the exact percentages of the granted degrees, but in the visualisation of the disproportion and changes, I can keep only the min and max value of the range (that is 0% and 100%).
